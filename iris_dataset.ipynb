```python
import numpy as np
from numpy import genfromtxt
from sklearn import linear_model

# commas seperated Iris dataset
file=genfromtxt('iris.csv',delimiter=',',dtype='str')
#print(file)

dic={}
count=0
for val in file:
    if val[5] not in dic:
        dic[val[5]]=count
        count+=1
print(dic)
for val in file:
    val[5]=dic[val[5]]
display(file)

#implement mechine Learning algroithim.
training_set=file[:130]
testing_set=file[130:]
#print(training_set.shape)
#print(testing_set.shape)

training_x=training_set[:,[1,2,3,4]]
training_x=training_x.astype(float)
training_y=training_set[:,[5]]

#print(training_x)
#print(training_y)

testing_x=testing_set[:,[1,2,3,4]]
testing_x=testing_x.astype(float)
testing_y=testing_set[:,[5]]
#print(testing_y)
#print(testing_x)

lr=linear_model.LogisticRegression()
lr.fit(training_x,training_y)

print(testing_x[1])
print(lr.predict([testing_x[1]]))
print('The REal value is:',testing_y[1])

#accuracy
print(lr.score(testing_x,testing_y)*100)
```
